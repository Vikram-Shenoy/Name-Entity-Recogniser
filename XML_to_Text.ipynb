# Converting all the training files to text ones

```python
import os
# assign directory
directory = 'sets/training-PHI-Gold-Set2'
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        tree = ET.parse(f)
        root = tree.getroot()
        for description in root.iter('TEXT'):
            a=description.text
            print(a)
        #open text file
        text_file = open("sets/training_text_file/all.txt", "a")
        #write string to file
        text_file.write(a)
        #close file
        text_file.close()
```

ALL XML training files have been added into one txt file
